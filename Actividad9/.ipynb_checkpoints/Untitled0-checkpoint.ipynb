```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import matplotlib as mpl
import math
mpl.use("pgf")
pgf_with_custom_preamble = {
    "font.family": "serif", # use serif/main font for text elements
    "text.usetex": True,    # use inline math for ticks
    "pgf.rcfonts": False,   # don't setup fonts from rc parameters
    "pgf.preamble": [
         "\\usepackage{units}",         # load additional packages
         "\\usepackage{unicode-math}",  # unicode math setup
         r"\setmathfont{xits-math.otf}",
         r"\setmainfont{DejaVu Serif}", # serif font via preamble
         ]
}
mpl.rcParams.update(pgf_with_custom_preamble)
from matplotlib.backends.backend_pdf import PdfPages


# Parametros
g = 9.806
l = 1.00
n = 200
radtodeg = 180.0/np.pi
epsilon = 0.001
# Define una serie de valores para el angulo inicial
theta_cero = np.linspace(epsilon, np.pi -epsilon, n) 

# Arreglos
integral     = [0 for i in range(n)]
integral0     = [0 for i in range(n)]
err         = [0 for i in range(n)]
perio_num     = [0 for i in range(n)]
sine         = [0 for i in range(n)]
erel0         = [0 for i in range(n)]
erel2         = [0 for i in range(n)]
erel4        = [0 for i in range(n)]
erel6        = [0 for i in range(n)]
erel8        = [0 for i in range(n)]
erel10        = [0 for i in range(n)]


# Definir la funcion a integrar: x - variable de int, y - param
# func = lambda x, k : 1.0 /(np.sqrt(np.cos(x)-np.cos(k)))


# Periodo aproximado con la ec. para angulos chicos
perio_teo = 2.0*np.pi*np.sqrt(l/g)

m = 0
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel0[j] = perio_num[j]/perio_teo 


m = 2
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel2[j] = perio_num[j]/perio_teo -1

m = 4
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel4[j] = perio_num[j]/perio_teo -2

m = 6
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel6[j] = perio_num[j]/perio_teo -3

m = 8
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel8[j] = perio_num[j]/perio_teo -4

m = 10
for i in range(0,m):
    for j in range(0,n):
        fac1 = float(math.factorial(2*(i)))
        fac2 = float((2**(i) * math.factorial(i))**2)
        sine[j] = np.sin(theta_cero[j]/2)**(2*(i))
        integral[j] = ((fac1/fac2)**2)*sine[j]
        integral0[j] = integral0[j] + integral[j]
        perio_num[j] = 2.0*np.pi*np.sqrt(l/g)*integral0[j]
        erel10[j] = perio_num[j]/perio_teo -5


# Convertir de radianes a grados para mostrar en el eje horizontal
x = radtodeg * theta_cero


# Graficar

with PdfPages('Act9.pdf') as pdf:
    plt.figure(figsize=(8, 8))
    plt.plot(x, erel0,'b', label=r"$T_{0}(\theta_{0})$")
    plt.plot(x, erel2,'g', label=r"$T_{2}(\theta_{0})$")
    plt.plot(x, erel4, 'r',label=r"$T_{4}(\theta_{0})$")
    plt.plot(x, erel6, 'c',label=r"$T_{6}(\theta_{0})$")
    plt.plot(x, erel8, 'm',label=r"$T_{8}(\theta_{0})$")
    plt.plot(x, erel10, 'y',label=r"$T_{10}(\theta_{0})$")
    plt.title("Error relativo del periodo utilizando serie de potencias.")
    plt.legend(loc='best')
    plt.xlabel("Theta0[grados]")
    plt.ylabel("Error relativo " r"$T/T_{0}$")
    plt.xlim([0, 90])
    plt.ylim([0, 1])
    plt.grid()
    pdf.savefig()  # saves the current figure into a pdf page
    plt.close()
    plt.show()
```
