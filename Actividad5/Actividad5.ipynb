```python
#theta'(t) = omega(t)
#omega'(t) = -b*omega(t) - c*sin(theta(t))

def pend(y, t, b, c):
     theta, omega = y
     dydt = [omega, -b*omega - c*np.sin(theta)]
     return dydt

b = 0.4 #fricción
g = 9.8 #gravedad
l = 4.5 #longitud de la cuerda
c = g/l

y0 = [np.pi/3, 0.9] #ángulo inical, velocidad angular

t = np.linspace(0, 20, 500) #rango de tiempo

from scipy.integrate import odeint
sol = odeint(pend, y0, t, args=(b, c))

import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='posicion(t)') #theta = posición
plt.plot(t, sol[:, 1], 'g', label='velocidad(t)') #omega = velocidad angular
plt.legend(loc='best')
plt.xlabel('Tiempo')
plt.grid()
plt.show()
```
