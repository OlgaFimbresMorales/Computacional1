```python
#theta'(t) = omega(t)
#omega'(t) = -b*omega(t) - c*sin(theta(t))

def pend(y, t, b, c):
     theta, omega = y
     dydt = [omega, -b*omega - c*np.sin(theta)]
     return dydt

b = 0.0 #fricción
c = 9.8 #g/l

y0 = [np.pi/2, 0.01] #ángulo, velocidad angular

t = np.linspace(0, 60, 500) #rango de tiempo

from scipy.integrate import odeint
sol = odeint(pend, y0, t, args=(b, c))

import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)') #theta = posición
plt.plot(t, sol[:, 1], 'g', label='omega(t)') #omega = velocidad angular
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
