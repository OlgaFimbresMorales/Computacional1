```python
# Modelo de apocalipsis zombie: 
# Modelo básico
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.ion()
plt.rcParams['figure.figsize'] = 10, 8

P = 0       # Tasa de Natalidad
d = 0.0001  # Muertes naturales por día
B = 0.0095  # Parámetro de transmisión  (per day)
G = 0.0001  # Removidos que se convierten en zombies (per day)
A = 0.0001  # Destruidos  (per day)

# solve the system dy/dt = f(y, t)
def f(y, t):
    Si = y[0] #suceptibles a convertirse
    Zi = y[1] #Zombies
    Ri = y[2] #removidos
    # Sistema de ecuaciones
    f0 = P - B*Si*Zi - d*Si #Suceptibles
    f1 = B*Si*Zi + G*Ri - A*Si*Zi #Zombies
    f2 = d*Si + A*Si*Zi - G*Ri #Removimos
    return [f0, f1, f2]

# Condiciones Iniciales
S0 = 500.                   # Pobación humana
Z0 = 0                      # Población zombie
R0 = 0                      # Población muerta
y0 = [S0, Z0, R0]   # initial condition vector
t  = np.linspace(0, 5., 1000)       # Tiempo

# Solución del sistema
soln = odeint(f, y0, t)
S = soln[:, 0]
Z = soln[:, 1]
R = soln[:, 2]

# plot results
plt.figure()
plt.plot(t, S, label='Humanos')
plt.plot(t, Z, 'r',label='Zombies')
plt.xlabel('Dias a partir del brote')
plt.ylabel('Poblacion')
plt.title('Apocalipsis Zombie: Modelo basico - Poblacion total viva; Sin nacimientos.')
plt.legend(loc=0)

# Cambio de condiciones iniciales
R0 = 0.01*S0   # 1% de población muerta
y0 = [S0, Z0, R0]

# Solución del sistema
soln = odeint(f, y0, t)
S = soln[:, 0]
Z = soln[:, 1]
R = soln[:, 2]

plt.figure()
plt.plot(t, S, label='Humanos')
plt.plot(t, Z, 'r',label='Zombies')
plt.xlabel('Dias a partir del brote')
plt.ylabel('Poblacion')
plt.title('Apocalipsis Zombie: Modelo basico - 1% Poblacion muerta; Sin nacimientos.')
plt.legend(loc=0)

# Cambio de condiciones iniciales
R0 = 0.01*S0   # 1% de población muerta
P  = 10        # 10 nacimientos por día
y0 = [S0, Z0, R0]

# Solución del sistema
soln = odeint(f, y0, t)
S = soln[:, 0]
Z = soln[:, 1]
R = soln[:, 2]

plt.figure()
plt.plot(t, S, label='Humanos')
plt.plot(t, Z, 'r',label='Zombies')
plt.xlabel('Dias a partir del brote')
plt.ylabel('Poblacion')
plt.title('Apocalipsis Zombie: Modelo basico - 1% Poblacion muerta; 10 nacimientos diarios')
plt.legend(loc=0)
```
